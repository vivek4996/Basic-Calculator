## **Histogrm to Understand the Distribuation**

```python
import warnings
warnings.filterwarnings("ignore")

for i in Merge_Dataset.select_dtypes(include="number").columns:
  sns.histplot(data=Merge_Dataset,x=i)
  plt.show()
```

## **BoxPlot to Identify Outliers**

```python
import warnings
warnings.filterwarnings("ignore")

for i in Merge_Dataset.select_dtypes(include="number").columns:
  sns.boxplot(data=Merge_Dataset,x=i)
  plt.show()
```

## **Correlation with Heatmap**

```python
S=Merge_Dataset.select_dtypes(include="number").corr()
```

```python
sns.heatmap(S,annot=True)
```
